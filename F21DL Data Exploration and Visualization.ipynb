## VISUALIZING DATA

```python
real_estate.plot(kind="scatter", x="YrSold", y="SalePrice")
```

<div class="alert alert-block alert-success">

```python
corr_matrix = real_estate.corr()
```

```python
corr_matrix["SalePrice"].sort_values(ascending = False)
```

```python
ax = sns.heatmap(corr_matrix)
#Heatmap showing the correlation of all attributes
```

```python
from pandas.plotting import scatter_matrix
attributes = ["SalePrice","Floor","N_APT", "N_Parkinglot"]
scatter_matrix(real_estate[attributes], figsize = (12,8))
```
